```python
#Steamer 2017 Hitter Projections

for url in Hsteamer_urls.values():
    text = BeautifulSoup(requests.get(url).text)
    players = text.find_all('tr', {'class': ['rgRow', 'rgAltRow']})
    for i in range(0, len(players)):
        names = players[i].find_all('a')
        teams = players[i].find_all('td')[2]
        plates = players[i].find_all('td')[4]
        wars = players[i].find_all('td')[27]
        [Hsteamer_player_lst.append(n.string) for n in names]
        [Hsteamer_team_lst.append(t.string) for t in teams]
        [Hsteamer_pa_lst.append(p.string) for p in plates]
        [Hsteamer_war_lst.append(w.string) for w in wars]
```

```python
#Steamer 2017 Pitcher Projections

for i in range(1,31):
    url = 'http://www.fangraphs.com/projections.aspx?pos=all&stats=pit&type=steamer&team=' + str(i) + '&lg=all&players=0'
    text = BeautifulSoup(requests.get(url).text)
    players = text.find_all('tr', {'class': ['rgRow', 'rgAltRow']})
    for i in range(0, len(players)):
        names = players[i].find_all('a')
        teams = players[i].find_all('td')[2]
        innings = players[i].find_all('td')[9]
        wars = players[i].find_all('td')[19]
        [Psteamer_player_lst.append(n.string) for n in names]
        [Psteamer_team_lst.append(t.string) for t in teams]
        [Psteamer_ip_lst.append(i.string) for i in innings]
        [Psteamer_war_lst.append(w.string) for w in wars]
```

```python
#2016 Season Hitter Performance

for k, v in H2016_urls.items():
    text = BeautifulSoup(requests.get(v).text)
    players = text.find_all('tr', {'class': ['rgRow', 'rgAltRow']})
    for i in range(0, len(players)):
        names = players[i].find_all('a')
        plates = players[i].find_all('td')[3]
        wars = players[i].find_all('td')[20]
        [H2016_player_lst.append(n.string) for n in names]
        H2016_team_lst.append(k)
        [H2016_pa_lst.append(p.string) for p in plates]
        [H2016_war_lst.append(w.string) for w in wars]
```

```python
#2016 Season Pitcher Performance

for k, v in P2016_urls.items():
    text = BeautifulSoup(requests.get(v).text)
    players = text.find_all('tr', {'class': ['rgRow', 'rgAltRow']})
    for i in range(0, len(players)):
        names = players[i].find_all('a')
        innings = players[i].find_all('td')[7]
        wars = players[i].find_all('td')[18]
        [P2016_player_lst.append(n.string) for n in names]
        P2016_team_lst.append(k)
        [P2016_ip_lst.append(i.string) for i in innings]
        [P2016_war_lst.append(w.string) for w in wars]
```

```python
h_steamer = pd.DataFrame([Hsteamer_player_lst, Hsteamer_team_lst, Hsteamer_pa_lst, Hsteamer_war_lst]).T
h_steamer.columns = ['player', 'team', 'pa', 'war']
h_steamer['2017_PA'] = h_steamer['pa'].astype(float)
h_steamer['2017_WAR_H'] = h_steamer['war'].astype(float)
```

```python
p_steamer = pd.DataFrame([Psteamer_player_lst, Psteamer_team_lst, Psteamer_ip_lst, Psteamer_war_lst]).T
p_steamer.columns = ['player', 'team', 'ip', 'war']
p_steamer['2017_IP'] = p_steamer['ip'].astype(float)
p_steamer['2017_WAR_P'] = p_steamer['war'].astype(float)
```

```python
h_2016 = pd.DataFrame([H2016_player_lst, H2016_team_lst, H2016_pa_lst, H2016_war_lst]).T
h_2016.columns = ['player', 'team', 'pa', 'war']
h_2016['pa'] = h_2016['pa'].astype(float)
h_2016['war'] = h_2016['war'].astype(float)
```

```python
p_2016 = pd.DataFrame([P2016_player_lst, P2016_team_lst, P2016_ip_lst, P2016_war_lst]).T
p_2016.columns = ['player', 'team', 'ip', 'war']
p_2016['ip'] = p_2016['ip'].astype(float)
p_2016['war'] = p_2016['war'].astype(float)
```

```python
#### Daily Bets
dailyBets_url = 'https://www.betdsi.eu/live-lines/baseball/major-league-baseball'
dailyBets_text = BeautifulSoup(requests.get(dailyBets_url).text)
```

```python
matchups = dailyBets_text.find_all('div', {'class': ['matchup']})
```

```python
homeTeam_lst = []
homeTeam_moneyLine = []
```

```python
#Home Teams
for i in matchups:
    home_teams = i.find_all('div', {'class': 'oddsD hTeam'})
    for h in home_teams:
        home_team = h.find_all('div', {'class': ['team']})
        line = h.find_all('div', {'class': ['money']})
        for t in home_team:
            v = t.find_all('span')
            for i in v:
                homeTeam_lst.append(i.string)
        for l in line:
            drill = l.find_all('span')
            for d in drill:
                homeTeam_moneyLine.append(d.string)
```

```python
homeTeam_lst = [x for x in homeTeam_lst if x != 'Action']
```

```python
homeTm_moneyLine = []
```

```python
for k, v in enumerate(homeTeam_moneyLine):
    if k in range(1,len(homeTeam_lst)*3, 3):
        homeTm_moneyLine.append(v.rstrip())
```

```python
homeTm_moneyLine
```

```python
awayTeam_lst = []
awayTeam_moneyLine = []
```

```python
#Away Teams
for i in matchups:
    away_teams = i.find_all('div', {'class': 'oddsD vTeam'})
    for a in away_teams:
        away_team = a.find_all('div', {'class': ['team']})
        line = a.find_all('div', {'class': ['money']})
        for t in away_team:
            v = t.find_all('span')
            for i in v:
                awayTeam_lst.append(i.string)
        for l in line:
            drill = l.find_all('span')
            for d in drill:
                awayTeam_moneyLine.append(d.string)
```

```python
awayTeam_lst = [x for x in awayTeam_lst if x != 'Action']
```

```python
awayTm_moneyLine = []
```

```python
for k, v in enumerate(awayTeam_moneyLine):
    if k in range(1,len(awayTeam_lst)*3, 3):
        awayTm_moneyLine.append(v.rstrip())
```
